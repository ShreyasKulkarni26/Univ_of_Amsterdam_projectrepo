```python
# %%
import glob
import os

def extract_rdepend_lines(ebuild_file):
    with open(ebuild_file, 'r', encoding='utf-8') as file:
        for line in file:
            if line.startswith("RDEPEND="):
                yield line.strip()
            elif line.startswith("DEPEND="):
                yield line.strip()
            elif line.startswith("BDEPEND="):
                yield line.strip()
            elif line.startswith("COMMON_DEPEND="):
                yield line.strip()
            elif line.startswith("USE_DEPEND="):
                yield line.strip()
            elif line.startswith("PREBUILT_DEPEND="):
                yield line.strip()
            elif line.startswith("CDEPEND="):
                yield line.strip()
            elif line.startswith("DBDEPEND="):
                yield line.strip()


def process_ebuild_files(folder_path):
    for root, directories, filenames in os.walk(folder_path):
        for filename in filenames:
            if filename.endswith(".ebuild"):
                filepath = os.path.join(root, filename)
                print(f"Processing ebuild file: {filepath}")
                rdepend_lines = extract_rdepend_lines(filepath)
                for line in rdepend_lines:
                    print(line) 
                    # Print extracted lines
                    # Further process the lines as needed

# Example usage:
root_folder = r"C:\Users\Shreyas kulkarni\Documents\Univ of Amsterdam\historical"  # Adjust as needed
process_ebuild_files(root_folder)


```
